```python
import numpy as np
import cv2
import matplotlib.pyplot as plt
%matplotlib inline
```

```python
img_raw = cv2.imread('./media/images/mandrill_colour.png')
if img_raw is None:
    print("Could not load the image")
```

```python
# Eso sucede porque OpenCV y matplotlib tienen diferentes órdenes de colores primarios.
# Mientras que OpenCV lee imágenes en forma de BGR, matplotlib, por otro lado, sigue el orden RGB

img_rgb = cv2.cvtColor(img_raw, cv2.COLOR_BGR2RGB)
plt.imshow(img_rgb)
```

```python
cv2.imwrite('final_image.png', img_rgb)
```
